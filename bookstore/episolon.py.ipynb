```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
##loading my individual bunny pictures
bunny1= cv2.imread('bunny/IMG_1.jpg')
bunny2= cv2.imread('bunny/IMG_2.jpg')
bunny3= cv2.imread('bunny/IMG_3.jpg')
bunny4= cv2.imread('bunny/IMG_4.jpg')
bunny5= cv2.imread('bunny/IMG_5.jpg')
bunny6= cv2.imread('bunny/IMG_6.jpg')
plt.imshow(bunny1)
```

```python
my_bunny_orig = cv2.cvtColor(bunny1, cv2.COLOR_BGR2RGB)
plt.imshow(my_bunny_orig)
```

```python
blend1_2 = cv2.addWeighted(src1=bunny2,alpha=0.5,src2=bunny1,beta=0.5,gamma=0)
#blend1_2_3 = cv2.addWeighted(src1=blend2_3,alpha=0.5,src2=bunny1,beta=0.5,gamma=0)
#blend3_4 = cv2.addWeighted(src1=bunny3,alpha=0.5,src2=bunny4,beta=0.5,gamma=0)
blend4_5 = cv2.addWeighted(src1=bunny4,alpha=0.5,src2=bunny5,beta=0.5,gamma=0)
final_blend = cv2.addWeighted(src1=blend1_2,alpha=0.5,src2=blend4_5,beta=0.5,gamma=0)
plt.imshow(final_blend)
```

```python
my_final_rgb_image =  cv2.cvtColor(final_blend, cv2.COLOR_BGR2RGB)
plt.imshow(my_final_rgb_image)
```

```python
gauss_filtered_image = cv2.GaussianBlur(my_final_rgb_image, (9, 9), 0)
plt.imshow(gauss_filtered_image)
```

```python
gaussImg1 = gauss_filtered_image[500:2450,:]
gaussImg2 = gauss_filtered_image[570:2520,:]
#gauss_filtered_image[500:,:]
# concat_test = np.concatenate((gaussImg1,gaussImg2), axis=1)
# plt.imshow(concat_test)
# my_final_rgb_image =  cv2.cvtColor(concat_test, cv2.COLOR_BGR2RGB)
# cv2.imwrite("FinalArtifact_test1.jpg", concat_test)
```

```python
concat_image = np.concatenate((gaussImg1,gaussImg2), axis=1)
plt.imshow(concat_image)
```

```python
my_final_rgb_image =  cv2.cvtColor(concat_image, cv2.COLOR_BGR2RGB)
cv2.imwrite("FinalArtifact.jpg", my_final_rgb_image)
```
